```python
import pandas as pd
import numpy
```

```python
filename="Recopilación coca cola.xlsx"

df=pd.read_excel(filename,1)
df['Trimestre']=df['Tiempo'].str[:1]
df['Tiempo']=df['Tiempo'].str[3:]
df.rename(columns={"Utilidad de operación " :"Utilidad_de_operacion","Ingresos totales":"Ingresos_totales","Utilidad bruta":"Utilidad_bruta","EBITDA Ajustado":"EBITDA"},inplace=True)
df=df[['Trimestre','Tiempo','Ingresos_totales','Utilidad_bruta','Utilidad_de_operacion','EBITDA']]
df=df.fillna("N/A")
print(df)

queryResultados="INSERT INTO RESULTADOS VALUES"

for idx in df.index:
    row=df.loc[idx]
    queryResultados+="({tiempo},{trimestre},{Ingresos},{UBruta},{UOperacion},{EBI},'Global'),\n".format(trimestre=row['Trimestre'],tiempo=row['Tiempo'],Ingresos=row['Ingresos_totales'],UBruta=row['Utilidad_bruta'],UOperacion=row['Utilidad_de_operacion'],EBI=row['EBITDA'])
    
print(queryResultados)
```
