```python
dfvolumen=pd.read_excel(filename,7)
#Columns ideales: País, trimestre, año, producto, cantidad
dfobjetivo=pd.DataFrame(columns=["trimestre","anio","producto","cantidad"])
transposedDf=dfvolumen.transpose()
#transposedDf=transposedDf.iloc[1:,:]
transposedDf[0]=transposedDf[0].str.replace(' (1)','')
transposedDf[0]=transposedDf[0].str.replace(' (2)','')
transposedDf[0]=transposedDf[0].str.replace('(1)','')
transposedDf[0]=transposedDf[0].str.replace('(2)','')
transposedDf.reset_index(inplace=True)

transposedDf.replace('Unnamed: [0-9][0-9][0-9]|Unnamed: [0-9][0-9]|Unnamed: [0-9]','NaN',regex=True,inplace=True)
#transposedDf['index']=transposedDf['index'].str[0:3]
rellenar=""
for idx in transposedDf.index:
    if transposedDf['index'].iloc[idx] == 'NaN':
        transposedDf.loc[idx,'index'] = rellenar
    else:
        rellenar=transposedDf['index'].iloc[idx]
        #print(rellenar)


dfobjetivo['producto']=transposedDf[0]
dfobjetivo['Tiempo']=transposedDf['index']
dfobjetivo['trimestre']=dfobjetivo['Tiempo'].str[0]
dfobjetivo['anio']=dfobjetivo['Tiempo'].str[3:]

dfmexico=dfobjetivo.copy()
dfmexico['cantidad']=transposedDf[1]
dfmexico=dfmexico.iloc[2:,:-1]
dfmexico.insert(0,"pais",'Mexico')
dfmexico.dropna

dfguatemala=dfobjetivo.copy()
dfguatemala['cantidad']=transposedDf[2]
dfguatemala=dfguatemala.iloc[2:,:-1]
dfguatemala.insert(0,"pais",'Guatemala')

dfcolombia=dfobjetivo.copy()
dfcolombia['cantidad']=transposedDf[5]
dfcolombia=dfcolombia.iloc[2:,:-1]
dfcolombia.insert(0,"pais",'Colombia')

dfbrasil=dfobjetivo.copy()
dfbrasil['cantidad']=transposedDf[6]
dfbrasil=dfbrasil.iloc[2:,:-1]
dfbrasil.insert(0,"pais",'Brasil')

dfargentina=dfobjetivo.copy()
dfargentina['cantidad']=transposedDf[7]
dfargentina=dfargentina.iloc[2:,:-1]
dfargentina.insert(0,"pais",'Argentina')

dfuruguay=dfobjetivo.copy()
dfuruguay['cantidad']=transposedDf[8]
dfuruguay=dfuruguay.iloc[2:,:-1]
dfuruguay.insert(0,"pais",'Uruguay')

dfFinal=pd.concat([dfmexico,dfguatemala,dfcolombia,dfbrasil,dfargentina,dfuruguay],ignore_index=True)

dfFinal.to_excel("exports/volumen.xlsx",index=False)
dfFinal.to_csv("mineria/volumen.csv",index=False)

print(dfFinal)
#print(transposedDf)
#dfobjetivo['producto']
#print(dfobjetivo)
#print(dfvolumen.columns)
#print(dfvolumen)
```
